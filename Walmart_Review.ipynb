```python
#Repeat 1000 times
```

```python
n = 300
iterations = 1000

female_purchase = df[df['Gender']=='F']['Purchase']
#female_purchase

female_sample_means = []
for i in range(iterations):
  sample_mean = female_purchase.sample(n,replace=True).mean()
  female_sample_means.append(sample_mean)

female_sample_means[:10]
```

```python
np.mean(female_sample_means)
```

```python
df.groupby('Gender')['Purchase'].describe()
```

```python
sns.displot(female_sample_means)
```

```python
np.mean(female_sample_means)
```

#Construction of interval

1. Z score method

2. Percentile method

```python
#Interval range = sample mean +- z*SE
```

```python
#95% CL
```

```python
from scipy.stats import norm
```

```python
norm.ppf(0.025), norm.ppf(0.975) #z_score
```

```python
np.mean(female_sample_means) - 1.96*np.std(female_sample_means)
```

```python
np.mean(female_sample_means) + 1.96*np.std(female_sample_means)
```

```python
female_upper_range = np.mean(female_sample_means) + 1.96*np.std(female_sample_means)/np.sqrt(n)
```

```python
female_lower_range = np.mean(female_sample_means) - 1.96*np.std(female_sample_means)/np.sqrt(n)
```

```python
female_lower_range, female_upper_range
```

```python
#Male

male_sample_means = [df[df['Gender']=='M']['Purchase'].sample(n,replace=True).mean() for i in range(iterations)]
```

```python
#male_sample_means
```

```python
np.mean(male_sample_means) - 1.96 * np.std(male_sample_means)
```

```python
np.mean(male_sample_means) + 1.96 * np.std(male_sample_means)
```

```python
male_upper_range = np.mean(male_sample_means) + 1.96 * np.std(male_sample_means)/np.sqrt(n)
male_lower_range = np.mean(male_sample_means) - 1.96 * np.std(male_sample_means)/np.sqrt(n)
```

```python
male_lower_range, male_upper_range
```

```python
#Percentile method
```

```python
female_range = np.percentile(female_sample_means,[2.5,97.5])
```

```python
male_range = np.percentile(male_sample_means,[2.5,97.5])
```

```python
print(female_range)
```

```python
print(male_range)
```
